```python
from random import randint
from time import sleep

def get_playlist_tracks(playlist_id): #queremos correr a funçao em qualquer id da playlist
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3000)/1000) # respectful nap
    return tracks
```

```python
all_tracks = get_playlist_tracks("5S8SJdl1BDc0ugpkEvFsIL")
len(all_tracks)
```

```python
from pandas import json_normalize
tracks2 = json_normalize(all_tracks)
```

```python
tracks2
```

```python
# we can even create a new column containing these DataFrames! KEEP song URI
# NOTE: need to change this function, since the columns names are now a bit different
def expand_list_dict2(row):  #pega numa row e olha para o track.artist e ve que é uma lista de dicionários
    df = json_normalize(row['track.artists'])
    df['song_id'] = row['track.id']
    return df
# criamos uma lova data frame artists_dfs onde tinha uma lista na coluna do track.artists
tracks2['artists_dfs'] = tracks2.apply(expand_list_dict2, axis=1)
tracks2['artists_dfs'][3]
```

```python
tracks2['artists_dfs'][6]
```

```python
# now we create a new dataframe with all these artists
artist_df2 = pd.DataFrame(columns=['href', 'id', 'name', 'type', 'uri'])
for mini_df in tracks2['artists_dfs']:
    #display(val)
    artist_df2 = pd.concat([artist_df2, mini_df], axis=0)
    
artist_df2
```

```python
# now we merge (join) the two dataframes, keeping only the fields we need
df_merged2 = pd.merge(left=tracks2,
                    right=artist_df2,
                    how='inner',
                    left_on='track.id',
                    right_on='song_id')
df_merged2.head()
```

```python
df_final2 = df_merged2[['track.name', 'name', 'song_id']]
df_final2
```

# Audio features

```python
chunks = [(i, i+100) for i in range(0, len(df_final2), 100)]
chunks  #separamos em bocados de 100
audio_features_list = []
for chunk in chunks:
    df_final2['song_id'] = df_final2['song_id'].astype(str)
    id_list100 = df_final2['song_id'][chunk[0]:chunk[1]] #esta a juntar os audios das musicas aos nomes das musicas (100 de cada vez)
    audio_features_list = audio_features_list + sp.audio_features(id_list100)
    sleep(randint(1,3000)/1000)
len(audio_features_list)
```

```python
audio_features_df = json_normalize(audio_features_list)
```

```python
audio_features_df.drop_duplicates(inplace=True) 
```

```python
df_w_audio_ft = pd.merge(left=df_final2,  #depois voltamos a fazer o merge do audio com a musica
                        right=audio_features_df,
                        how='inner',
                        left_on='song_id',
                        right_on='id')
df_w_audio_ft
```
